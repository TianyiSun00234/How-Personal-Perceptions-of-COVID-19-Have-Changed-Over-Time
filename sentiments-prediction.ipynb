```python
train_dataset = df.anxiety_num[-107:-50]    ##############################change 
val_dataset = df.anxiety_num[-55:]        ##############################change
```

```python
# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
	X= list()
	for i in range(len(sequence)):
		# find the end of this pattern
		end_ix = i + n_steps
		# check if we are beyond the sequence
		if end_ix > len(sequence)-1:
			break
		# gather input and output parts of the pattern
		seq_x = sequence[i:end_ix]
		X.append(seq_x)
	return np.array(X)
X = split_sequence(train_dataset, 3)
y = split_sequence(val_dataset, 2)
```

```python
# multi-step encoder-decoder lstm example
from numpy import array
from keras.models import Sequential
from keras.layers import LSTM
from keras.layers import Dense
from keras.layers import RepeatVector
from keras.layers import TimeDistributed

# reshape from [samples, timesteps] into [samples, timesteps, features]
X = X.reshape((X.shape[0], X.shape[1], 1))
y = y.reshape((y.shape[0], y.shape[1], 1))
# define model
model = Sequential()
model.add(LSTM(100, activation='relu', input_shape=(3, 1)))
model.add(RepeatVector(2))
model.add(LSTM(100, activation='relu', return_sequences=True))
model.add(TimeDistributed(Dense(1)))
model.compile(optimizer='adam', loss='mse')
# fit model
model.fit(X, y, epochs=100, verbose=0)
```

```python
# demonstrate prediction
x_input = array(split_sequence(list(df.anxiety_num), 3))          ##############################change
x_input = x_input.reshape((x_input.shape[0], x_input.shape[1], 1))
yhat = model.predict(x_input, verbose=0)
#print(yhat)
```
